```python
import xgboost as xgb
import pandas as pd 
import numpy as np 
import lightgbm as lgb
from lightgbm.sklearn import LGBMRegressor
from sklearn.model_selection import KFold, train_test_split, GridSearchCV
import matplotlib.pyplot as plt
from Dataloader import Dataloader
from score import score
from sklearn.utils import shuffle
from IPython.display import display, HTML
from datetime import date
%matplotlib inline
```

```python
model = xgb.XGBRegressor(colsample_bytree=0.4,
                objective = "reg:linear",
                 gamma=0,                 
                 learning_rate=0.1,
                 n_estimators=100,                                                                    
                 subsample=0.6,
                 seed=42,
                 verbosity = 0)
clf = GridSearchCV(model,
                   {'max_depth': [8,16],
                    'colsample_bytree': [0.4,0.6,0.8],
                    'reg_alpha' : [0,5,10],
                    'reg_lambda': []
                   }, verbose=0)
```

```python
#model = xgb.XGBRegressor()
```

```python
df = pd.read_csv("data/train.csv")
dataloader = Dataloader()
X_train, X_test, y_train, y_test = dataloader.prepare_train_data(df,one_hot=False)
```

```python
def evalerror(y_preds,dtrain):
    labels = dtrain.get_label()
    return 'error', np.sum(np.abs(y_preds-labels))/len(y_preds)

xgb_model = model.fit(X_train, y_train,
                     eval_metric=evalerror,
                     eval_set = [(X_train, y_train),(X_test,y_test)],
                     early_stopping_rounds = 100)
```

```python
predictions = xgb_model.predict(X_train)
actuals = y_train
predictions = predictions*X_train['building_area']
actuals = actuals*X_train['building_area']
print(score(actuals, predictions))
```

```python
predictions = xgb_model.predict(X_test)
actuals = y_test
predictions = predictions*X_test['building_area']
actuals = actuals*X_test['building_area']
print(score(actuals, predictions))
```

```python
test_df = pd.read_csv("data/test.csv")
build_id = pd.read_csv("data/submit_test.csv")['building_id']
test_df = dataloader.prepare_test_data(test_df)
test_y = model.predict(test_df)
submit_df = pd.DataFrame(data={
    'building_id' : build_id,
    'total_price' : test_y*test_df['building_area']
})
today = str(date.today())
submit_df.to_csv(f'submits/{today}_xgboost.csv',index = False)
```
